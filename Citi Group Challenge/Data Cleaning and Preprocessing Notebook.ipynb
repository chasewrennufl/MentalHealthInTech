```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# format for floats
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df_survey = pd.read_csv('survey.csv', index_col='Timestamp')

df_survey.tail()
```

```python
df_survey = df_survey.drop('comments', axis=1)
```

```python
df_tech_survey = df_survey[df_survey.tech_company != 'No']
```

```python
df_tech_survey.iloc[:,: 12]
```

```python
df_tech_survey.drop(df_tech_survey[df_tech_survey['Age'] >= 90].index, inplace = True)
```

```python
df_tech_survey.drop(df_tech_survey[df_tech_survey['Age'] < 18].index, inplace = True)
```

```python
binary_genders = ['Male', 'Female']
df_tech_survey.loc[df_tech_survey['Gender'].str.lower().str.startswith('f'), 'Gender'] = 'Female'
df_tech_survey.loc[df_tech_survey['Gender'].str.lower().str.startswith('m'), 'Gender'] = 'Male'
df_tech_survey.loc[~df_tech_survey['Gender'].isin(binary_genders), 'Gender'] = 'Other'
```

```python
df_tech_survey.drop('no_employees', axis=1)
```

```python
csv_data = df_tech_survey.to_csv('Clean Mental Health Survey Data.csv', index=True)
```
